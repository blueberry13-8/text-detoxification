```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("s-nlp/t5-paranmt-detox")
model = AutoModelForSeq2SeqLM.from_pretrained("s-nlp/t5-paranmt-detox")
```

```python
examples = [
    # references
    ['shit', 'fuck', 'i pay to watch them masturbate', 'drop the fucking gun',
     'I have orders to kill her.', "you're becoming disgusting."],
    # translations
    ['blimey!', 'Shoot', 'sometimes i only pay for looking', 'drop the gun',
     "I've got orders to put her down.", "Now you're getting nasty."],
]
```

```python
for i in range(len(examples[0])):
    input_ids = tokenizer(
        examples[0][i], return_tensors="pt"
    ).input_ids

    output = model.generate(input_ids, max_new_tokens=50)

    print(f'Example #{i + 1}')
    print(f'Input:       {examples[0][i]}')
    print(f'Output:      {tokenizer.decode(output[0], skip_special_tokens=True)}')
    print(f'Translation: {examples[1][i]}')
    print()
```
